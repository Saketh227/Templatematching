```python
import matplotlib.pyplot as plt
```

```python
import cv2 as cv
import numpy as np
img=cv.imread('img_1.png',0)
temp=cv.imread('temp_10.png',0) # matching

res=cv.matchTemplate(img,temp,cv.TM_CCOEFF_NORMED)

cv.imshow('image',img)
cv.imshow('template',temp)
cv.waitKey(0)
cv.destroyAllWindows
```

```python
print(plt.imshow(img))
```

```python
print(plt.imshow(temp))
```

```python
loc=np.where(res>=0.877)

if loc[1].size!=0:
    w,h=temp.shape[::-1]
    
    min_val, max_val, min_loc, max_loc = cv.minMaxLoc(res)
    top_left = max_loc
    bottom_right = (top_left[0] + w, top_left[1] + h)
    cv.rectangle(img,top_left, bottom_right,(0,255,0),5)
    cv.imshow('Match is found',img)
    cv.waitKey(0)
    cv.destroyAllWindows()
else:
    cv.imshow('no match is found in the image',img)
    cv.waitKey(0)
    cv.destroyAllWindows()
```

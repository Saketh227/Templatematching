```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

img=cv.imread('img_1.png')

grey_img=cv.cvtColor(img,cv.COLOR_BGR2GRAY)

template=cv.imread('temp_10.png',0)# loads in gray scale

res=cv.matchTemplate(grey_img,template,cv.TM_CCOEFF_NORMED)

# print(res)

loc=np.where(res>=0.877)

print(loc)

w,h=template.shape[::-1]

for pt in zip(*loc[::-1]):
    cv.rectangle(img,pt,(pt[0]+w,pt[1]+h),(0,255,0),5)

plt.imshow(img)
```

```python
cv.imshow('options',img)
cv.imshow('sample',template)

cv.waitKey(0)

cv.destroyAllWindows()
```

```python
import cv2 as cv
import numpy as np


img=cv.imread('img_1.png')

grey_img=cv.cvtColor(img,cv.COLOR_BGR2GRAY)

template=cv.imread('temp_10.png',0)# loads in gray scale

res=cv.matchTemplate(grey_img,template,cv.TM_CCOEFF_NORMED)


min_val, max_val, min_loc, max_loc= cv.minMaxLoc(res)

# print(min_val)

# print(max_val)

# print(min_loc)

print(max_loc)

height,width = template.shape[:2] # shape() gives o/p height ,width
print(height,width)

top_left= max_loc

bottom_right= (top_left[0] + width, top_left[1] + height)

cv.rectangle(img, top_left, bottom_right, (0,0,255),3)

plt.imshow(img)
```
